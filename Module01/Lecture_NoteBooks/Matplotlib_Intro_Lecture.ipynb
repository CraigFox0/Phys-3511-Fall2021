# Matplotlib Intro NoteBook

When importing matplotlib, we can use Jupyter magics to have plots displayed inline. We have 2 options for this
- inline: static plots
- notebook: interactive plots

If we do not use magics, then we will need to do
> plt.show()

to display the plot as though it were a python only code

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib notebook
#%matplotlib inline
```

```python
x = np.arange(0,100)
y = x**2
```

Two ways to plot these
- Functional Method (review on own)
- Object Oriented Method **(perferred)**

# Matplotlib Object Oriented Method

In this approach we will instanciate a figure objects and call methods and attributes off of them.
Typically best to keep all commands related to a fit in the same cell

We can now add axes to the object fig, which defines the canvas area we will draw on. 
To add axes we use the add_axes method
> fig.add_axes([lower_left, bottom left, width, height])

All arguments are from 0 to 1

So to plot:
- Create empty fig object
- Add axes
- plot off the axes

Note: semi-colon suppresses output text from the figure

```python
fig = plt.figure()
axes = fig.add_axes([0.15,0.1,0.8,0.8])
axes.plot(x,y)
axes.set_xlabel('x')
axes.set_ylabel('y')
axes.set_title('y vs. x')
```

Plotting 2 figures on the same canvas

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.16,0.45,0.4,0.4])
axes1.plot(x,y)
axes2.plot(y,x)
axes1.set_title('Large Plot')
axes2.set_title('Small Plot');
#stop print out after plot with ;
```

# Subplots
axes object is just an array, so we can iterate therough them and manipulate them via indexing just like we do for arrays.
We can use the subplots method
>fig = plt.subplots(nrows, ncols)

```python
#tuple unpack
fig, axes = plt.subplots(1,2)

axes[0].plot(x,y)
axes[1].plot(y,x)
axes[0].set_title('first')
axes[0].set_xlabel('x')
axes[0].set_ylabel('y')
axes[1].set_title('second')
axes[1].set_xlabel('x')
axes[1].set_ylabel('y')
plt.tight_layout();
```

# Multiple Curves on the Same Graph

```python
fig = plt.figure()
ax = fig.add_axes([0.1,0.1,0.8,0.8])
ax.plot(x,y,label='$x^2$')
ax.plot(x,x**2.5,label='$x^{2.5}$')
ax.legend();
```

# Different Plot Types
- Scatter
- 1D Histograms
- 2D Histograms

```python
new_y = np.random.randn(1000)
fig = plt.figure()
ax = fig.add_axes([0.1,0.1,0.8,0.8])
ax.set_xlabel('x')
ax.set_title('Hist')
ax.hist(new_y,bins=100);
```

```python
fig = plt.figure()
ax = fig.add_axes([0.1,0.1,0.8,0.8])
ax.scatter(x,y,marker='o')
ax.set_xlabel('x')
ax.set_title('Scatter')
```
